### Get a list of all the cars in a particular Bonhams auction
* Here I went to the 2019 Bonhams Monterey auction and clicked all at the bottom of the page to find the url that loads all the cars for this aution.
* It appears that you may add the following query string to any auction to load all of the cars: /#/aa0=1&MR0_length=224&w0=list&m0=0'
* For example https://www.bonhams.com/auctions/ + Auction Number + /#/aa0=1&MR0_length=224&w0=list&m0=0 will load all the cars for a particular bonhams auction.

```python
#This URL is the same as going to the aution main page and clicking the "all" link at the bottom 
url = 'https://www.bonhams.com/auctions/25221/#/aa0=1&MR0_length=224&w0=list&m0=0'
getAllCarsInAuction(url)
```
